```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Creates a DataFrame that displays how many unique players purchased items in the data
total_players = pd.DataFrame({'Total Players':[len(purchase_data['SN'].unique())]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Creates a DataFrame that summarizes the total purchases of the dataset
purchase_data['Item Name'].unique()
purchase_summary = pd.DataFrame({'Number of Unique Items': [len(purchase_data['Item ID'].unique())], 'Average Price': [purchase_data['Price'].mean()], 'Number of Purchases': [len(purchase_data['Price'])], 'Total Revenue': [purchase_data['Price'].sum()]})
purchase_summary['Average Price'] = purchase_summary['Average Price'].map("${:.2f}".format)
purchase_summary['Total Revenue'] = purchase_summary['Total Revenue'].map("${:.2f}".format)
purchase_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Creates a new copy of the data that only includes each person once for demographic analysis.
people_data = purchase_data.drop_duplicates('SN')
people_data.count()
```

```python
# Creates a Dataframe that counts the number of instances of each of the three genders in the dataset.
people_data['Gender'].value_counts()
dict = {'Total Count': {}}
for item, index in zip(people_data['Gender'].value_counts(), people_data['Gender'].value_counts().index):
    dict['Total Count'][index] = item
gender_demographics = pd.DataFrame(dict)
gender_demographics
```

```python
# Calculates the percentage each gender makes up of the dataset
total = gender_demographics['Total Count'].sum()
percentages = []
for x in range(0,3):
    percentages.append(((gender_demographics.iloc[x,0]/total)*100).round(2))

gender_demographics['Percentage of Players'] = percentages
gender_demographics = gender_demographics.sort_values(by = 'Total Count', ascending = False)
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase data analysis based on gender
gender_purchase = purchase_data.groupby(['Gender'])['Price'].agg(['count', 'mean', 'sum'])
gender_purchase = gender_purchase.sort_values(by = 'count', ascending = False)
avg_total = []
for x in range(0,3):
    avg_total.append((gender_purchase.iloc[x, 2]/gender_demographics.iloc[x, 0]).round(2))
gender_purchase['Avg Total Purchase Per Person'] = avg_total
gender_purchase['mean'] = gender_purchase['mean'].map("${:.2f}".format) 
gender_purchase['sum'] = gender_purchase['sum'].map("${:.2f}".format)
gender_purchase['Avg Total Purchase Per Person'] = gender_purchase['Avg Total Purchase Per Person'].map("${:.2f}".format)
gender_purchase = gender_purchase.rename(columns = {'count': 'Purchase Count', 'mean': 'Average Purchase Price', 'sum':'Total Purchase Value'})

gender_purchase



```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# DataFrame displaying the age demographics of the dataset
bins = [0,9,14,19,24,29,34,39,total_players.iloc[0,0]]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_demographics = pd.DataFrame()
age_demographics['Total Count'] = pd.cut(people_data['Age'], bins, labels = labels)
#age_demographics = age_demographics.groupby(['Group']).agg('count')
age_demographics = age_demographics['Total Count'].value_counts().to_frame()

age_percentages = []
for x in range(0, len(age_demographics['Total Count'])):
    age_percentages.append(((age_demographics.iloc[x,0]/age_demographics['Total Count'].sum())*100).round(2))
age_demographics['Percentage of Players'] = age_percentages
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase data analysis based on the age groups of the purchasers
age_purchase = pd.DataFrame()
age_purchase = purchase_data.loc[:,['Age', 'Price']]
age_purchase['Purchases'] = pd.cut(purchase_data['Age'], bins, labels = labels)
age_purchase['Purchase Count'] = age_purchase['Purchases'].value_counts().to_frame()
age_purchase = age_purchase.drop(columns = 'Purchase Count')
age_purchase = age_purchase.groupby(['Purchases'])['Price'].agg(['count', 'mean', 'sum'])
age_purchase = age_purchase.sort_values(by = 'count', ascending = False)

avg_age_purchase = []
for x in range(0, len(age_purchase['count'])):
    avg_age_purchase.append(((age_purchase.iloc[x,2]/age_demographics.iloc[x,0])).round(2))
    
age_purchase['Avg Total Purchase per Person'] = avg_age_purchase
age_purchase['mean'] = age_purchase['mean'].map("${:.2f}".format)
age_purchase['sum'] = age_purchase['sum'].map("${:.2f}".format)
age_purchase['Avg Total Purchase per Person'] = age_purchase['Avg Total Purchase per Person'].map("${:.2f}".format)
age_purchase = age_purchase.rename(columns = {'count': 'Purchase Count', 'mean': 'Average Purchase Price', 'sum': 'Total Purchase Value'})


age_purchase
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Purchase data analysis DataFrame showing the individuals who spent in the dataset
top_spenders = pd.DataFrame()
top_spenders = purchase_data.groupby(['SN'])['Price'].agg(['count', 'mean', 'sum'])
top_spenders = top_spenders.sort_values(by = 'sum', ascending = False)
top_spenders = top_spenders.head()
top_spenders = top_spenders.rename(columns = {'count': 'Purchase Count', 'mean':'Avg Purchase Price', 'sum':'Total Purchase Value'})
top_spenders['Avg Purchase Price'] = top_spenders['Avg Purchase Price'].map("${:.2f}".format)
top_spenders['Total Purchase Value'] = top_spenders['Total Purchase Value'].map("${:.2f}".format)
top_spenders




```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# DataFrame of the most popular items purchased
popular_items = pd.DataFrame()
popular_items = purchase_data.groupby(['Item Name', 'Item ID'])['Price'].agg(['count', 'mean', 'sum'])

popular_items = popular_items.sort_values(by = ['count', 'mean'], ascending = False)
popular_items = popular_items.rename(columns = {'count':'Purchase Count', 'mean': 'Item Price' ,'sum': 'Total Purchase Value'})

popular_items['Item Price'] = popular_items['Item Price'].map("${:.2f}".format) 
popular_items['Total Purchase Value'] = popular_items['Total Purchase Value'].map("${:.2f}".format) 
popular_items.head(10)


```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# DataFrame showing the most profitable items in the dataset
profitable_items = pd.DataFrame()
profitable_items = purchase_data.groupby(['Item Name', 'Item ID'])['Price'].agg(['count', 'mean', 'sum'])
profitable_items = profitable_items.sort_values(by = 'sum', ascending = False)
profitable_items = profitable_items.rename(columns = {'count':'Purchase Count', 'mean': 'Item Price' ,'sum': 'Total Purchase Value'})

profitable_items['Item Price'] = profitable_items['Item Price'].map("${:.2f}".format) 
profitable_items['Total Purchase Value'] = profitable_items['Total Purchase Value'].map("${:.2f}".format) 
profitable_items.head(10)
```
